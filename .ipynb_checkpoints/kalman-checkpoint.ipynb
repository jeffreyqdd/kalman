```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
'''
creating basic simulation with noise
'''

#altitude_current = 0.98 * altitude_previous

initial_height = 1000      #only touch this
noise = 40                #+/- unit

#--------------------------logic

altitude_current = -1
altitude_previous = initial_height

y = [altitude_previous]
x = [0]

for i in range(1,500):
    altitude_current = altitude_previous * 0.98
    
    y.append(altitude_current + np.random.uniform(-noise, noise))
    x.append(i)
    
    altitude_previous = altitude_current
    
plt.plot(x,y)
plt.show()
```

```python
'''
now we have two equations describing the state of our airplane
    1) altitude_current = 0.98 * altitude_previous <-- true state
    2) observed_altitude_current = altitude_current + noise_current <-- observed state

if we generalize these two equations then:
    * x_k = a * x_k-1 <-- current state equals previous state * change
    * z_k = x_k + v_k <-- current measurement equals current state + noise
    
we can make things a tad rougher with turbulence
    x_k = ax_k-1 + w_k
    
'''




#altitude_current = 0.98 * altitude_previous

initial_height = 1000      #only touch this
noise = 40                #+/- unit
turbulance = 5             #+/1 unit

#--------------------------logic

altitude_current = -1
altitude_previous = initial_height

y_state = [altitude_previous + np.random.uniform(-turbulance, turbulance)]
y_meas = [altitude_previous]

x = [0]


for i in range(1,500):
    altitude_current = altitude_previous * 0.98 + np.random.uniform(-turbulance, turbulance)
    
    y_meas.append(altitude_current + np.random.uniform(-noise, noise))
    y_state.append(altitude_current)
    
    x.append(i)

    altitude_previous = altitude_current
    
plt.plot(x,y_meas, label = "measurement")
plt.plot(x,y_state, label = "state")
plt.legend()
plt.show()
```

```python
'''
given these two equations 
    * x_k = a * x_k-1 + w_k   or state = previous_state + process_noise
    * z_k = x_k + v_k         or measurement = state + measurement_noise
    
since we're trying to obtain state from measurements
    * x_k = z_k - v_k         or state = measurement - measurement_nois

unfortunately, we do not know noise, however, we can make estimates via the gain, or kalman gain (g_k)
    * x_hat_k = x_hat_k-1 + g_k (z_k - x_hat_k-1)
    * or estimated state = estimated previous state + gain * (measurement - estimated previous state)
    * if gain = 0, then our new state = previous state...we disregard measurement
    * if gain - 1, then new state = previous measurement...we disregard previous state.
'''


gain = 0.4



#--------------------------logic
y_predicted = [initial_height]
for measurement in y_meas[1:]:
    predicted = y_predicted[-1] + gain * (measurement - y_predicted[-1])
    y_predicted.append(predicted)

    
plt.scatter(x,y_meas, label = "measurement",s=1)
plt.plot(x,y_state, label = "state")
plt.plot(x,y_predicted, label = "predicted location")
plt.legend()
plt.show()    
```
